**Dataset Exploration**

```python
import pandas as pd
```

```python
import os
from pathlib import Path

# Walk up from CWD until we find the project root (contains Dataset/Raw/neo.csv)
# Works regardless of where the Jupyter server was launched from
_search = Path(os.getcwd())
for _ in range(6):
    if (_search / 'Dataset' / 'Raw' / 'neo.csv').exists():
        break
    _search = _search.parent
else:
    raise FileNotFoundError(
        "Could not locate neo.csv. Make sure the Jupyter server is started "
        "from within the MLOPS-v2 project directory."
    )

DATASET_PATH = str(_search / 'Dataset' / 'Raw' / 'neo.csv')
df = pd.read_csv(DATASET_PATH)
print(f"Loaded dataset from: {DATASET_PATH}")
```
